```python
# import necessary libraries
import gradio as gr
import cv2
from ultralytics import YOLO
import torch
from fastrtc import WebRTC, get_hf_turn_credentials 
```

```python
# Load YOLOv11 pose model
model = YOLO("yolo11n-pose.pt")
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
print(device)

# Load credentials for Hugging Face TURN server
TOKEN = "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" 
creds = get_hf_turn_credentials(token=TOKEN)
```

```python
css = """.my-group {max-width: 600px !important; max-height: 600px !important;}
         .my-column {display: flex !important; justify-content: center !important; align-items: center !important;}"""

def process(image):
    image = cv2.flip(image, 1)
    results = model(image)
    result = results[0].plot(boxes = False)  # Overlay keypoints on the image
    return result

with gr.Blocks(css=css) as demo:
    gr.HTML(
        """
        <h1 style='text-align: center'>
        CoE 197 Z - Machine Exercise 2: Real-time Human Body Keypoint Tracking
        </h1>
        <h5 style='text-align: center'>
        This is a keypoint-tracking app using Gradio which utilizes the yolo11n-pose model <br> John Eric Estrada <br> SN 2022-02217
        </h5>
        """
    )
    with gr.Column(elem_classes=["my-column"]):
        with gr.Group(elem_classes=["my-group"]):
            image = WebRTC(label="Stream", rtc_configuration=creds, mode='send-receive')

        image.stream(
            fn=process, inputs=[image], outputs=[image]
        )
```
